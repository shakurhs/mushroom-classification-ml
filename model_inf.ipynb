```python
# import libraries
import pandas as pd
import pickle
```

```python
# loading best model dari file pickle
with open('best_rf_model.pkl', 'rb') as model_file:
    model = pickle.load(model_file)
```

```python
# membuat data inference yang akan diprediksi
df_inf = {
    'cap_diameter' : 2.70,
    'cap_shape' : 'convex',
    'cap_surface' : 'grooves',
    'cap_color' : 'red',
    'does_bruise_or_bleed' : 'no',
    'gill_attachment' : 'adnate',
    'gill_spacing' : 'close',
    'gill_color' : 'white',
    'stem_height' : 7.15,
    'stem_width' : 2.71,
    'stem_color' : 'brown',
    'has_ring' : 'no',
    'ring_type' : 'none',
    'habitat' : 'woods',
    'season' : 'summer'
}

# menyimpan data inference ke dalam bentuk tabel
df_inf = pd.DataFrame([df_inf])
df_inf
```

```python
# prediksi data inference 
y_pred_inf = model.predict(df_inf)

if y_pred_inf == 0:
    print('Hasil prediksi model inference adalah:', y_pred_inf, 'dan termasuk kategori Edible Mushroom')
else:
    print('Hasil prediksi model inference adalah:', y_pred_inf, 'dan termasuk kategori Inedible/Poisonous Mushroom')
```
